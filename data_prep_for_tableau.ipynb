# Goal

the 'genre' column in the 'IBSN_only.xlsx' dataset contains a string of genres separated by commas for each row in teh dataset. The goal is to get a unique list of all genres found in the dataset to create a table that shows how many times a genre appears in the dataset. That dataset will then be used in Tableau to visualize the most prominent genres in the dataset.

```python
df = pd.read_excel('ISBN_only.xlsx')
```

```python
df.head()
```

```python
#want to fill nan values in dataset so that it doesn't mess with future code
df.fillna("",inplace = True)
#Add a comma at the end of every values in the genre for when I combine all the values in the column into one string.
df['genre'] = df['genre'] + ','
```

```python
#want to concat all the strings into one string to then get a unique list of all genres in this dataset.

big_string = ''
for i in df['genre']:
    big_string = big_string + str(i)
```

```python
unique_genres = list(set(big_string.split(',')))
```

```python
len(unique_genres)
```

```python
#this dictionary will hold all the unique genres in the data with the key being the genre and the value being
#how many times that genre shows up in the dataset.
d = {}
for i in unique_genres:
    counter = 0
    for b in df['genre']:
        if i in str(b):
            counter += 1
    d[i] = counter
    
#sorting the dictionary by how many times the genre appears in the d ataset
d = {k: v for k, v in sorted(d.items(), key=lambda item: item[1],reverse = True)}


```

```python
#dataframe of sorted dictionary with genre and count column
genres = pd.DataFrame(list(d.items()), columns=['Genre', 'Count'])
genres = genres.loc[genres['Genre'] != '']
#I'll use this dataframe to create a visualization in tableau
genres
```

'genre_count.xlsx' will be used in Tableau to create a bar chart showing the top N genres in the dataset.

```python
genres.to_excel('genre_count.xlsx', index = False)
```
